###Model Building

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss = []

for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, init="k-means++", max_iter=300, random_state = 42)
  kmeans.fit(data)

  wcss.append(kmeans.inertia_)
```
