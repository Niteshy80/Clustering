```python
#PCA
#1) standardize the data

from sklearn.preprocessing import StandardScaler
```

```python
SS = StandardScaler()

data_scaled = SS.fit_transform(data)
```

```python
print(data_scaled)
```

```python
#2) Coveriance Matrix

cov_mat = np.cov(data_scaled.T)
cov_mat
```

```python
#3) Eigenvalues and eigenvectors

eig_val, eig_vec = np.linalg.eig(cov_mat)
print ('EigenValues ', eig_val)
print ('EigenVectors ', eig_vec)
```

```python
#4) Explained Variance ration

total = sum(eig_val)
total
```

```python
var_exp = [(i/total)*100 for i in sorted (eig_val, reverse = True)]
cum_sum = np.cumsum(var_exp)
cum_sum
```

```python
plt.figure(figsize = (15,10))
plt.plot(cum_sum)
plt.xlabel("Components")
plt.ylabel("Variable_ratio")
plt.title("graph")
plt.show()
```

```python
from sklearn.cluster import KMeans

wcss = []

for cluster in range (1,12):
    kmean = KMeans (n_clusters = cluster, init = 'k-means++', random_state=2)
    kmean.fit_predict(data)
    
    wcss.append(kmean.inertia_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
clusters_range = np.arange(2,11)
```

```python
for cluster in clusters_range:
 kmeans = KMeans(n_clusters=cluster,init='k-means++',max_iter=300,n_init=10)
 kmeans.fit_predict(data)
 
 score = silhouette_score(data, kmeans.labels_, metric='euclidean')
 print(f'for cluster: {cluster} --> Silhouetter Score: %.3f' % score)
```

## Model 2 Hierarichal clustering

```python
from sklearn.cluster import AgglomerativeClustering

sil_list = []

for cluster in range (2,11):
    for linkage_method in ['single', 'ward', 'average', 'complete']:
        aglo = AgglomerativeClustering(linkage=linkage_method, affinity='euclidean', n_clusters=cluster).fit_predict(data)
        
        sil_score = silhouette_score(data, aglo, metric='euclidean')
        sil_list.append((cluster, sil_score, linkage_method, len(set(aglo)) ) )

df_hierarchical = pd.DataFrame(sil_list, columns=['clusters', 'sil_score','linkage_method', 'number_of_clusters'])
```

```python
df_hierarchical.sort_values(sil_score, assending = False)
```
